* Importar el primero
* Bucle para ir recorriendo los ficheros (importante tener una consistencia en el nombre)
    * Importamos el fichero
    * Lo apendizamos al final del primero

```python
filepath = "../datasets/distributed-data/"

data = pd.read_csv(filepath + "001.csv")
length = len(data)

for i in range(2,333):
    if i<10:
        filename = "00" + str(i)
    elif i<100:
        filename = "0" + str(i)
    else:
        filename = str(i)
    file = filepath + filename + ".csv"
    next_data = pd.read_csv(file)
    length += len(next_data)
    data = pd.concat([data,], axis = 0)
```
