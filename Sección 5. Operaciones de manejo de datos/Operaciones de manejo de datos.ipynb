```python
import pandas as pd
```

```python
raw_data = pd.read_csv('../datasets/exercises/downloaded_covid_effects.csv')
raw_data.head()
```

## Selección de columnas

```python
column_names = raw_data.columns.values.tolist()
column_names
```

```python
column_names = list(set(column_names)-set(['Unnamed: 0'])) #listas a conjuntos para restarlas y de nuevo a lista
column_names
#alternativa: seleccionar directamente
#column_names = ['Direction', 'Year', 'Date', 'Weekday', 'Current_Match', 'Country', 'Commodity', 'Transport_Mode', 'Measure', 'Value', 'Cumulative']
```

```python
data = raw_data[column_names]
data.head()
```
