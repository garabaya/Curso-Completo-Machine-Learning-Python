## La simulaci√≥n de Montecarlo

```python
# Generar dos n√∫meros aleatorios entre 0 y 1 ( x e y)
# Calcular x*x + y*y (ecuaci√≥n del c√≠rculo)
    # Si el valor es inferior a 1 estoy dentro del c√≠rculo
# N¬∫ de puntos dentro / n¬∫ intentos = aproximaci√≥n de pi
# Repetir el experimento un n¬∫ suficiente de veces (100)
# Calcular el promedio de las 1000 aproximaciones

pi_avg = 0
pi_value_list = []
n = 1000 # n¬∫ de puntos
n2 = 100 # n¬∫ de veces
for i in range(n2):
    value = 0
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()
    for j in range(n):
        z = np.sqrt(x[j]*x[j] + y[j]*y[j])
        if z<=1:
            value +=1
    float_value = float(value)
    pi_value = float_value *4 / n
    pi_value_list.append(pi_value)
    pi_avg += pi_value
    
pi = pi_avg/n2
print(pi)
plt.hist(pi_value_list)
```

```python
def pi_montecarlo(n,n2):
    pi_avg = 0
    pi_value_list = []
    n = 1000 # n¬∫ de puntos
    n2 = 100 # n¬∫ de veces
    for i in range(n2):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j]*x[j] + y[j]*y[j])
            if z<=1:
                value +=1
        float_value = float(value)
        pi_value = float_value *4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value

    pi = pi_avg/n2
    print(pi)
    #figure = plt.plot(pi_value_list)
    figure, axes = plt.subplots()
    axes.add_artist(plt.Circle((0.5,0.5),0.5,alpha=.3))
    axes.set_aspect(1)
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.scatter(x,y,10,color="r")
    return (pi, figure)
```

```python
pi_montecarlo(1000000000,2000000)
```
